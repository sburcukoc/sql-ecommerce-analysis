```python
df = pd.read_excel("supermarket_sales_analysis.xlsx", sheet_name=" supermarket_Sales", header=1)

df.head()
```

```python
df["Total"].sum()
```

```python
df.groupby("City")["Total"].sum()
```

```python
df.groupby("City")["Total"].sum().plot(kind="bar")

plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("City")["Total"].sum().plot(kind="bar")

plt.title("Total Sales by City")
plt.xlabel("City")
plt.ylabel("Total Sales")
plt.show()
```

```python
df.groupby("Product line")["Total"].sum().sort_values(ascending=False).plot(kind="bar")

plt.title("Total Sales by Product Line")
plt.xlabel("Product Line")
plt.ylabel("Total Sales")
plt.show()
```
